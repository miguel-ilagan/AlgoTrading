```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-bright")
```

```python
# Compare performance of closing price of respective stocks
# Normalise the values by dividing by the starting closing price
normalised_close=close.div(close.iloc[0])
normalised_close.plot(figsize=(15,8),fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
# Calculate percentage change of stocks (daily)
daily_return = close.pct_change().dropna()
daily_return
```

```python
# Plot distribution of percentage return of selected stocks
for column in daily_return.columns:
    fig, ax = plt.subplots(figsize=(12,8))
    daily_return[column].plot(kind="hist",figsize=(12,8), bins=100)

    ax.set_xlabel("Percentage return")
    ax.set_ylabel("Frequency")
    ax.set_title(f"Percentage return of {column}") 
    plt.show()
```

```python
# Summary statistics of the selected stocks
business_days_in_a_year = 252
daily_stats = daily_return.describe().T.loc[:,["mean","std"]]
annual_stats = daily_stats.copy()

annual_stats["mean"] = annual_stats["mean"] * business_days_in_a_year
annual_stats["std"] = annual_stats["std"] * np.sqrt(business_days_in_a_year)
```

```python
# Plot the annual statistics
annual_stats.plot.scatter(x="std",y="mean",figsize=(12,8),s=50,fontsize=15)
for i in annual_stats.index:
    plt.annotate(i,xy=(annual_stats.loc[i,"std"]+0.002,annual_stats.loc[i,"mean"]+0.002),size=15)
plt.xlabel("Annual risk(std)",fontsize = 15)
plt.ylabel("Annual return",fontsize = 15)
plt.title("Risk/return",fontsize=25)
plt.show()
```

```python
moving_averages = close.copy()
for column in close.columns:
    moving_averages[f'{column} SMA50'] = close[column].rolling(window=50,min_periods=50).mean()
    moving_averages[f'{column} SMA200'] = close[column].rolling(window=200,min_periods=200).mean()

moving_averages.dropna(inplace=True)
```

```python
moving_averages
```

```python
# Plot price with its moving averages

for column in close.columns:
    moving_averages[[f'{column}', f'{column} SMA50', f'{column} SMA200']].plot(figsize=(15,8),fontsize=12)
    plt.legend(fontsize=12)
    plt.show()
```

```python
# Cummulative returns and drawdowns
crad = close.copy()
for column in crad.columns:
    crad[f'{column} logreturns'] = np.log(crad[column].div(crad[column].shift(1)))
    crad[f'{column} cummreturns'] = crad[f'{column} logreturns'].cumsum().apply(np.exp)
    crad[f'{column} cummax'] = crad[f'{column} cummreturns'].cummax()
    crad[f'{column} drawdown'] = crad[f'{column} cummax'] - crad[f'{column} cummreturns']
    crad[f'{column} drawdown %'] = crad[f'{column} drawdown'] / crad[f'{column} cummax']

crad.dropna(inplace=True)
```

```python
for column in crad.columns:
    crad[[f'{column} cummreturns',f'{column} cummax', f'{column} drawdown %']].plot(figsize=(12,8),title=f"{column} buy and hold + cummmax + drawdown %", 
                fontsize=12, secondary_y=f'{column} drawdown %')
    plt.show()
```

```python
class SMABacktester():
    def __init__(self,symbol,SMA_S, SMA_L,start,end):
        self.symbol=symbol
        self.SMA_S=SMA_S
        self.SMA_L=SMA_L
        self.start=start
        self.end=end
        self.results=None
        self.get_data()
        
    def get_data(self):
        df=yf.download(self.symbol,start=self.start,end=self.end)
        data=df.Close.to_frame()
        data["returns"]=np.log(data.Close.div(data.Close.shift(1)))
        data["SMA_S"]=data.Close.rolling(self.SMA_S).mean()
        data["SMA_L"]=data.Close.rolling(self.SMA_L).mean()
        data.dropna(inplace=True)
        self.data2=data
        
        return data
    
    def test_results(self):
        data=self.data2.copy().dropna()
        data["position"]=np.where(data["SMA_S"]>data["SMA_L"],1,-1)
        data["strategy"]=data["returns"]*data.position.shift(1)
        data.dropna(inplace=True)
        data["returnsbh"]=data["returns"].cumsum().apply(np.exp)
        data["returnsstrategy"]=data["strategy"].cumsum().apply(np.exp)
        perf=data["returnsstrategy"].iloc[-1] # this chooses the final cumulative return for the SMA strategy
        outperf=perf-data["returnsbh"].iloc[-1] # this chooses the final cumulative return for the buy and hold strategy and calculates the difference with the SMA strategy
        self.results=data

        # Cumulative returns for the SMA strategy
        ret=np.exp(data["strategy"].sum())
        std= data["strategy"].std()*np.sqrt(252)
    
        #return ret,std
        return round(perf,6), round(outperf,6)
    
    def plot_results(self):
        if self.results is None:
            print("Run the test please")
        else:
            title="{}| SMA_S={} | SMA_L{}".format(self.symbol,self.SMA_S, self.SMA_L)
            self.results[["returnsbh","returnsstrategy"]].plot(title=title, figsize=(12,8))
```
